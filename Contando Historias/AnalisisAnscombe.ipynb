#### Graphs in Statistical Analysis
El problema con los datos $\textit{extraos}$ es que lo ms fcil de hacer es borrarlos, pero, en que radica un buen anlisis?, en la diversidad de nuestra informacin, si tenemos un monton de datos iguales y hacemos una regrecin lneal de estos, en realidad no estaramos exprimiendole informacin a nuestros datos, de esta manera pues no tendra sentido hacer un anlisis si no tenemos variedad.\
El problema no son los datos, es que no se manejan de la manera adecuada, la razn de que las estadsticas sean iguales y la regresin salga de la misma manera es que no se han manejado los datos de la manera adecuada, existen maneras de hacerlo, una de ellas es estandarizar los datos, de esta manera no se ignora la diversidad de nuestros datos y los datos atpicos no afectan demasiado, esto se debe a que de esa manera se le resta peso a esos datos.\
Otro problema es que en muchas ocasiones se manejan conjuntos de datos como si fueran funciones, cuando en realidad no lo son, recordando el criterio de la recta vertical, parafraceando: si al grficar los datos al dibujar una recta, esta recta toca ms de un punto, entonces no es una funcin, es decir, es una funcin s y solo s la recta toca un solo punto, es por ello que si se hubiesen graficado los datos antes de hacer cualquier anlisis para ver el comportamiento, se habran dado cuenta.
