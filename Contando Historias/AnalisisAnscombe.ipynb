#### Graphs in Statistical Analysis
El problema con los datos $\textit{extraños}$ es que lo màs fàcil de hacer es borrarlos, pero, ¿en que radica un buen anàlisis?, en la diversidad de nuestra informaciòn, si tenemos un monton de datos iguales y hacemos una regreciòn lìneal de estos, en realidad no estarìamos exprimiendole informaciòn a nuestros datos, de esta manera pues no tendrìa sentido hacer un anàlisis si no tenemos variedad.\
El problema no son los datos, es que no se manejan de la manera adecuada, la razòn de que las estadìsticas sean iguales y la regresiòn salga de la misma manera es que no se han manejado los datos de la manera adecuada, existen maneras de hacerlo, una de ellas es estandarizar los datos, de esta manera no se ignora la diversidad de nuestros datos y los datos atìpicos no afectan demasiado, esto se debe a que de esa manera se le resta peso a esos datos.\
Otro problema es que en muchas ocasiones se manejan conjuntos de datos como si fueran funciones, cuando en realidad no lo son, recordando el criterio de la recta vertical, parafraceando: si al gràficar los datos al dibujar una recta, esta recta toca màs de un punto, entonces no es una funciòn, es decir, es una funciòn sì y solo sì la recta toca un solo punto, es por ello que si se hubiesen graficado los datos antes de hacer cualquier anàlisis para ver el comportamiento, se habrìan dado cuenta. Ahì la importancia de los gràficos en un anàlisis.
